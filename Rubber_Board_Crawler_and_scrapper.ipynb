# Importing Libraries

```python
import os
import pandas as pd
import numpy as np
import datetime
import time
import dateutil
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup as bs
```

# Setting up dates in form

```python
#setting up the time to and from date

now = datetime.datetime.today()
from_date = now + dateutil.relativedelta.relativedelta(months = -2)
from_date = from_date.strftime("%d-%m-%Y")
to_date = now + dateutil.relativedelta.relativedelta(months = -1)
to_date = to_date.strftime("%d-%m-%Y")
```

# Crawler and browse parameters

```python
#setting up web browser parameter
options = Options()
options.add_argument("--start-maximized")
#options.add_argument("--headless")
driver = webdriver.Chrome(options = options)


#crawler
driver.get("http://www.rubberboard.org.in/public")
time.sleep(3)  #page loading delay

#clicking on National daily Link on the home page
driver.find_element_by_link_text("Daily").click()
time.sleep(15) #page laoding delay

#filling from date field
datefieldfrom = driver.find_element_by_xpath('//input[@id="txtFromDate"]')  
datefieldfrom.click()
datefieldfrom.send_keys(from_date)  #from_date

#filling to date field
datefieldto = driver.find_element_by_id('txtToDate')
datefieldto.click()
datefieldto.send_keys(to_date)  #to_date

#selecting GRADE from the select list
driver.find_element_by_xpath('//select[@id="grade"]/option[@value="10"]').click()  #isnr20 grade


time.sleep(7)
#clicking on GO Button
driver.find_element_by_xpath('//button[@class="btn btn-info pull-right btn-block btn-sm"]').click()  #go_button
time.sleep(15) # page loading delay

#getting the required data from web driver
data = driver.page_source
```

# Data Scrapping

```python
from bs4 import BeautifulSoup as bs
```

```python
#HTML Parsing in BeautifulSpoup

soup = bs(data,'html.parser') #beautifulSoup HTML Parser
div = soup.find('div', {'id' : 'Kottayam'}) #Finding the required "Kottayam" div from the HTML page
table_rows = div.findAll('tr') #finding the required table rows of data set
```

```python
#Extracting data from table rows

filtered_data = []
for row in table_rows :
    cols = row.findAll('td')
    cols = [ele.text.strip() for ele in cols]
    filtered_data.append(cols)

filtered_data
```

# Creating dataframe in pandas

```python
cols = ['Date', 'Price in RS', 'Change in RS', 'Price in Dollars', 'Change in Dollars']
Kottayam_isnr20_national_data = pd.DataFrame(data = filtered_data, columns = cols)
```
